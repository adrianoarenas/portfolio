```python
def transform_to_web_mercator(row):
    transformer = Transformer.from_crs("epsg:4326", "epsg:3857")
    return (transformer.transform(row.latitude, row.longitude))
```

```python
def apply_web_mercantor(df):
    transformation = df.apply(transform_to_web_mercator, axis=1)
    transformation_df = transformation.apply(pd.Series)
    transformation_df.columns=['longitude_wm','latitude_wm']
    return transformation_df
```

```python
f, ax = plt.subplots(1, figsize=(8, 9))
ax.scatter('longitude', 'latitude', data = marine_reserve_coords)
ax.scatter([-90.70716011821771],[-0.1990199429494878], marker = 'x', s = 100)
```

```python
marine_coords_ext = []
for index, row in marine_reserve_coords.iterrows():
    B = (-0.1990199429494878, -90.70716011821771) #Point B (lat, lon) - Center
    A = (row.latitude, row.longitude)
    s = -40 #Distance (km)
    
    #Define the ellipsoid
    geod = Geodesic.WGS84
    
    #Solve the Inverse problem
    inv = geod.Inverse(A[0],A[1],B[0],B[1])
    azi1 = inv['azi1']
    print('Initial Azimuth from A to B = ' + str(azi1))
    
    #Solve the Direct problem
    dir = geod.Direct(A[0],A[1],azi1,s*1609)
    C = (dir['lat2'],dir['lon2'])
    marine_coords_ext.append(C)
    print('C = ' + str(C))
```

```python
marine_reserve_ext_df = pd.DataFrame(marine_coords_ext, columns = ['latitude', 'longitude'])
```

```python
marine_reserve_ext_df = normalise_column_names(marine_reserve_ext_df, 'latitude', 'longitude')
marine_reserve_ext_wm = apply_web_mercantor(marine_reserve_ext_df)
```

```python
ref_angle = (int(sample[0]['HEADING']) + 90)%360
dist = 500000
```

```python
if 0 <= ref_angle < 90:
    cos_quadrant = 1
    sin_quadrant = 1
elif 90 <= ref_angle < 180:
    cos_quadrant = -1
    sin_quadrant = 1
elif 180 <= ref_angle < 270:
    cos_quadrant = -1
    sin_quadrant = -1
else:
    cos_quadrant = 1
    sin_quadrant = -1
```

New Coords

```python
angle_rad = math.radians(ref_angle)
```

```python
lon2 = lon1 + dist * math.cos(angle_rad)*cos_quadrant
lat2 = lat1 + dist * math.sin(angle_rad)*sin_quadrant
```
