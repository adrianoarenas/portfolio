# Approximating pi by Monte Carlo

Given we have a square of side D and a circle of diameter D:

A_circle = pi * (D/2)^2 = pi * (D^2)/4

A_square = D^2

If we replace D^2 in A_circle with A_square and solve for pi, we get:

pi = 4*(A_cirlcle/A_square)

I will use a quarter of a circle for ease of calculations, but results should be the same as its proportional.

```python
import random
import math
import matplotlib.pyplot as plt
```

```python
def MC_pi(iterations):
    points_circle = 0
    points_total = 0
    plot_circle_x = []
    plot_circle_y = []
    plot_square_x = []
    plot_square_y = []
    for i in range(iterations):
        x = random.random()
        y = random.random()
        distance = math.sqrt(x**2 + y**2)
        if distance < 1:
            points_circle += 1
            plot_circle_x.append(x)
            plot_circle_y.append(y)
        else:
            plot_square_x.append(x)
            plot_square_y.append(y)
    
    return((float(points_circle)/iterations)*4, plot_circle_x, plot_circle_y, plot_square_x, plot_square_y)
```

```python
pi, circle_x, circle_y, square_x, square_y = MC_pi(5000)
print("With 5000 iterations, pi: ", pi)

fig, ax = plt.subplots(figsize=(8,8))
ax.set_aspect('equal')
ax.scatter(circle_x, circle_y, color='g', marker='o')
ax.scatter(square_x, square_y, color='r', marker='o')
```

```python
pi, circle_x, circle_y, square_x, square_y = MC_pi(50000)
print("With 50000 iterations, pi: ", pi)

fig, ax = plt.subplots(figsize=(8,8))
ax.set_aspect('equal')
ax.scatter(circle_x, circle_y, color='g', marker='o')
ax.scatter(square_x, square_y, color='r', marker='o')
```
